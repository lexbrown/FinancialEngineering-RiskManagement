```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def compound_interest(pv, r, t):
    fv = pv * (1+r)**t
    return fv
```

```python
def pv_annuity(pv, t, r, method='year', due='post'):
    if method == 'year':
        annuity_payment = pv / (1/r - 1/(r*(1+r)**t))
        return annuity_payment
    if method == 'month':
        annuity_payment = pv / (1/(r/12) - 1/((r/12)*(1+r/12)**(t*12)))
        return annuity_payment
```

```python
def irr(cf, n, face, investment):
    r = 10 #лучше стартовать с 1000%
    #сумма n членов геом прогресссии и остаточная стоимость:
    value = cf * (((1/(1+r))**n-1)/-r) + face/((1+r)**n) - investment
    while abs(value) > 0.1: #пока PV сильно отклоняется от 0
        if value < 0:
            step = r/2
            r = r - step
        if value > 0:
            r = r + step/2
            step = step/2
        value = cf * (((1/(1+r))**n-1)/-r) + face/((1+r)**n) - investment
    return r
```

```python
def draw_option(p, strike, op_type='call', how='long'):
    import matplotlib.pyplot as plt
    %matplotlib inline
    if op_type == 'call' and how == 'long':   
        gain = [-p, -p, strike*2-p] #oy
        stock_price = [0, strike, strike*2] #ox
    if op_type == 'call' and how == 'short':   
        gain = [p, p, strike*(-2)+p] #oy
        stock_price = [0, strike, strike*2] #ox
    plt.figure(figsize=(8,5))
    plt.plot(stock_price, [0]*len(stock_price), color='black')
    plt.plot(stock_price, gain, color='blue', lw=2)
    plt.grid()
```

```python
draw_option(30, 200, how='short')
```

```python
def option_gain(stock_price, p, strike, op_type='call', how='long'):
    if op_type =='call' and how =='long':
        return max(-p, stock_price-strike-p)
    if op_type =='call' and how =='short':
        return min(p, -stock_price+strike+p)

assert option_gain(120, 40, 110) == -30
assert option_gain(50, 30, 110, how='short') == 30
```
